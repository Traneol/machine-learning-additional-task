 установим пакеты зависимости и разберёмся с модулями:

```python
import time
import keras_cv
from tensorflow import keras
import matplotlib.pyplot as plt
```

Сначала построим модель:

```python
model = keras_cv.models.StableDiffusion(img_width=512, img_height=512)
```

Затем создадим текстовое описание. Например, «фото человека паука на бегемоте»

```python
images = model.text_to_image("photo of spider-man sitting on a hippo", batch_size=3)


def plot_images(images):
    plt.figure(figsize=(20, 20))
    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i])
        plt.axis("off")


plot_images(images)
```
